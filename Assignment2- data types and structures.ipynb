14. How does the "in" keyword work differently for lists and dictionaries?

Ans. The **in** keyword in python is used to check if a value is exits within a collection . However its behaviour differs between lists and dictionaries due to the structure of these data types.
1. **in for Lists**:
In a list , the in keyword checks whether a specific element exists in the list by iterating through the entire list and comparing each element to the one you are looking for.

```python
my_list=[1,2,3,4,5]
print(3 in my_list)
```

2. **in** for dictionaries:  In dictionaries , the in keyword behaves differently. It checks whether a key exists in the dictionary not the value. You are checking for the presence of the key itself, not the associated value.

```python
my_dict={"apple":5,"banana":10}
print("apple"in my_dict)
```

17. Describe the time complexity of accessing elements in a dictionary.

Ans. In Python, dictionaries are implemented using hash table which makes access to elements very efficient. Lets break down the time complexity of various dictionary , particulary access.

1. **Average Case(O(1))**: When you access an element by key, Python uses the hash value of the key to directly find the corresponding value in the dictionary. This operation is typically done in constant time, regardless of the number of elements in the dictionaries.
2. **Worst Case(O(n))**: In some rare cases, the time complexity of accessing an element can degrade  to O(n). This happens when many dictionary key hash to the same value, causing the hash to fall back on a slower method of resolving the collisions.

19. Why are dictionaries considered unordered and how does that affect data retrieval?

Ans. Dictionaries in programming lanuages are considered as unordered collection because they store key value pair based on a hashing algorithm rather than in a specific sequence. This means that the order in which you insert items is not guaranteed to be the same  when you retrieve them . The primary reason for this unordered nature is to optimize for quick lookups, insertion and deletion.

**How this Affect data retrival**
1. Efficiency: Since dictionaries use hash tables, they provide average time complexity of O(1) for lookups, insertions and deletions make them very efficient for these operations.
2. No order: You cannot rely on ordered of items in dictionary. If you maintain order, you should use **ORDEREDDICT**.
3. Iteration: When iterating over a dictioanary the order of items is not predictable.

20. Explain the difference between list and a dictionaries terms of data retreival.

Ans.**LISTS**:
1. **Ordered Collection**: Lists are maintain the order of elements . Items are retrieved based on their position on the list.
2. **Index Based Access**: You retrieve items using index .The time complexity of accessing  an item is O(1)
